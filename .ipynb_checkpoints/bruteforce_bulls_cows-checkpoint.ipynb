```python
def calculate_zero(guess, lst):
    lst = [num for num in lst if num[0] != guess[0] or num[1] != guess[1] or num[2] != guess[2] or num[3] != guess[3]]
    return lst


def calculate_one(guess, lst):
    lst1 = [num for num in lst if
            num[0] == guess[0] and num[1] != guess[1] and num[2] != guess[2] and num[3] != guess[3]]
    lst2 = [num for num in lst if
            num[0] != guess[0] and num[1] == guess[1] and num[2] != guess[2] and num[3] != guess[3]]
    lst3 = [num for num in lst if
            num[0] != guess[0] and num[1] != guess[1] and num[2] == guess[2] and num[3] != guess[3]]
    lst4 = [num for num in lst if
            num[0] != guess[0] and num[1] != guess[1] and num[2] != guess[2] and num[3] == guess[3]]
    lst = set(lst1 + lst2 + lst3 + lst4)
    return lst


def calculate_two(guess, lst):
    lst1 = [num for num in lst if
            num[0] == guess[0] and num[1] == guess[1] and num[2] != guess[2] and num[3] != guess[3]]
    lst2 = [num for num in lst if
            num[0] == guess[0] and num[1] != guess[1] and num[2] == guess[2] and num[3] != guess[3]]
    lst3 = [num for num in lst if
            num[0] == guess[0] and num[1] != guess[1] and num[2] != guess[2] and num[3] == guess[3]]

    lst4 = [num for num in lst if
            num[0] != guess[0] and num[1] == guess[1] and num[2] == guess[2] and num[3] != guess[3]]
    lst5 = [num for num in lst if
            num[0] != guess[0] and num[1] == guess[1] and num[2] != guess[2] and num[3] == guess[3]]

    lst6 = [num for num in lst if
            num[0] != guess[0] and num[1] != guess[1] and num[2] == guess[2] and num[3] == guess[3]]
    lst = set(lst1 + lst2 + lst3 + lst4 + lst5 + lst6)
    return lst


def calculate_three(guess, lst):
    lst1 = [num for num in lst if
            num[0] != guess[0] and num[1] == guess[1] and num[2] == guess[2] and num[3] == guess[3]]
    lst2 = [num for num in lst if
            num[0] == guess[0] and num[1] != guess[1] and num[2] == guess[2] and num[3] == guess[3]]
    lst3 = [num for num in lst if
            num[0] == guess[0] and num[1] == guess[1] and num[2] != guess[2] and num[3] == guess[3]]
    lst4 = [num for num in lst if
            num[0] == guess[0] and num[1] == guess[1] and num[2] == guess[2] and num[3] != guess[3]]
    lst = set(lst1 + lst2 + lst3 + lst4)
    return lst


lst = [f'{num:04d}'for num in range(0, 10000)]

guess_lst = []
n = int(input())
for _ in range(n):
    guess, g_correct = input().split()
    guess_lst.append([guess, g_correct])

for g in guess_lst:
    if int(g[1]) == 0:
        lst = calculate_zero(g[0], lst)
    elif int(g[1]) == 1:
        lst = calculate_one(g[0], lst)
    elif int(g[1]) == 2:
        lst = calculate_two(g[0], lst)
    elif int(g[1]) == 3:
        lst = calculate_three(g[0], lst)
else:
    print(list(lst)[0])
```
