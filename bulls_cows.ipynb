```python
from random import choice

def check(ans,exp): #check 函数，可被其他脚本使用
    correct = 0
    present = 0
    exp = list(exp)
    for i in range(len(ans)):
        if ans[i] == exp[i]:
            correct += 1
            exp[i] = 'X'
    for i in range(len(ans)):
        if ans[i] in exp[:i]+exp[i:]:
            present += 1 
    return '{}A{}B'.format(correct,present)

def calculate(init_guess, ab_val, exp): # calculate 函数，可被其他脚本使用
    lst = [num for num in (num for num in (map(lambda x: '{:04d}'.format(x),range(0,10000))) if '0' not in num)\
           if num[0] not in num[1:] and num[1] not in num[0] + num[2:] and num[2] not in num[0:2] + num[3] and num[3] not in num[0:3]]
    possible = [num for num in lst if check(num,exp) in [ab_val,'4A0B'] and check(init_guess,num)[0] == ab_val[0]]
    return possible
    
if __name__ == '__main__': # 游戏只能在直接跑 guess.py 的时候运行
    lst = [num for num in (num for num in (map(lambda x: '{:04d}'.format(x),range(0,10000))) if '0' not in num)\
           if num[0] not in num[1:] and num[1] not in num[0] + num[2:] and num[2] not in num[0:2] + num[3] and num[3] not in num[0:3]]
    expected = choice(lst)
    player_guess_bank = set()
    ai_guess_bank = set()
    ai_guess = choice(lst)
    best_rating = 0
    best_result = '0A0B'
    best_guess = '1234'
    while True:
#        print('\n测试所以给你答案: {}'.format(expected)) # 完成就删
        player_guess = input("请输入你的猜测: ")
        player_guess_bank.add(player_guess)
        player_result = check(player_guess,expected)
        print('您的猜测结果: {}'.format(player_result))
        if player_result == '4A0B':
            print('猜对了，您胜出了！游戏结束！')
            break
        current_rating = int(player_result[0])*4 + int(player_result[2])
        if current_rating >= best_rating:
            best_rating = current_rating
            best_guess = player_guess
            best_result = player_result
        while True:
            ai_guess = choice(calculate(best_guess, best_result, expected))
            if ai_guess in (player_guess_bank|ai_guess_bank):
                continue
            else:
                break
        result = check(ai_guess,expected)
        ai_guess_bank.add(ai_guess)
        current_rating = int(result[0])*4 + int(result[2])
        if current_rating >= best_rating:
            best_rating = current_rating
            best_guess = ai_guess
            best_result = result
        print('电脑的猜测：{}'.format(ai_guess))
        print('电脑的猜测结果: {}'.format(result))
        if result == '4A0B':
            print('抱歉，电脑胜出！游戏结束！')
            break
```
